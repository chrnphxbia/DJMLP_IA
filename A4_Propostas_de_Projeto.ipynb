# **Resumo**

O **objetivo** deste projeto é **desenvolver um modelo de classificação de sentimentos (positivo, negativo e neutro) em tweets relacionados à invasão dos Três Poderes em 8 de janeiro de 2023**. A pesquisa utiliza como base o dataset **Tweet_Eleições_2022** (SILVA et al., 2024), composto por aproximadamente 9,5 milhões de tweets coletados durante o período eleitoral brasileiro de 2022 através da API oficial do *Twitter* (Twitter Developer Platform, 2022).

Embora o dataset esteja disponível publicamente em versão **desidratada** (com remoção de conteúdo textual e metadados sensíveis para preservação de privacidade), este estudo conta com acesso privilegiado aos dados hidratados completos através da participação de uma das integrantes no projeto Heiwa (Projeto de pesquisa do grupo CIBERDEM). Essa condição permite análises textuais detalhadas que não seriam possíveis com a versão pública. Os dados passaram por pré-processamento que incluiu: (1) remoção de duplicatas, (2) filtragem temporal para o recorte específico de 8 de janeiro de 2023, e (3) anonimização de usuários seguindo os princípios éticos do projeto Heiwa e *Twitter*.

Para o desenvolvimento da solução, serão utilizadas as bibliotecas scikit-learn, spaCy e NLTK.
